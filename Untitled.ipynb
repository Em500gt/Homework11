```python
from sympy import *
x = Symbol('x')
f = -12*x**4*sin(cos(x))-18*x**3+5*x**2+10*x-30
print(f)
```

```python
plot(f, (x, -100, 100))
```

```python
def fuk(start, stop, s):
    a = start
    b = stop
    d = 10
    for _ in range(18):
        a *= 10
        b *= 10
        d *= 10

        for i in range(a, b):
            param = i / d
            if f.subs(x, param) < 0 and s < 0:
                a = i - 1
                b = i + 1
                break
            elif f.subs(x, param) > 0 and s > 0:
                a = i - 1
                b = i + 1
                break
    print(f"{a / d, b / d} - > {f.subs(x, (a / d + b / d) / 2)}")

first = f.subs(x, -10.0)
for i in range(-99, 100):
    param = i /10
    second = f.subs(x, param)
    if first > 0 and second < 0 or second > 0 and first < 0:
        start = (i - 1) / 10
        stop = param
        fuk(int(start * 10), int(stop * 10), second)
        first = second
```
